```python
#Importing the required libraries
import pandas as pd
import numpy as np

#Let us load the dataset from the data folder
# The ".." means to go up one folder level
# Then '/data/' goes into the data folder

df = pd.read_csv('../data/Mall_Customers.csv')

#Let us display the first 5 rows of the dataset
df.head()
```

```python
#Let us explore the dataset to understand its structure and contents
#Checking how many rows and columns we have
print("Dataset shape:", df.shape)
print("Rows:", df.shape[0], "| Columns:", df.shape[1])
```

```python
#Show all column names
print("Column names:", df.columns.tolist())
```

```python
# Check the datatypes for each column
print("Print datatypes:")
print(df.dtypes)
```

```python
# Let us check for any missing values in the dataset
print("Missing values in each column:")
print(df.isnull().sum())
```

```python
# Let us check for duplicates in the dataset
print("Number of duplicate rows:", df.duplicated().sum())
```

```python
# Let us get statistical summary of the dataset
print("Statistical summary of the dataset:")
print(df.describe())
```

```python
# Checking for unique values in the gender column
print("Unique genders:", df['Gender'].unique())
print("Count of each gender:")
print(df['Gender'].value_counts())
```
